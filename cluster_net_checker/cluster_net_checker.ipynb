# 网络工具demo
## 功能

1. 网络拓扑展示
    - 支持万卡+拓扑展示（选择并展示集群中特定通信域拓扑）
    - 通信域信息准确（从训练中采集）
    - 物理拓扑获取（调研中）

2. 问题rank分析
    - 快慢卡
    - 慢链路
    - rank指标排序


## 使用流程

采集+分析+可视化

1. 采集：
    - 采集会借助profiling工具，并在mindspeed中额外插入代码来获取并行信息
    - profiling工具动态采集
    - 插入到mindspeed的patch代码（获取并行信息，并使用mstx将并行信息写入profiling db）

2. 分析：
    - cluster_net_checker.py

3. 可视化：
    - 快速穿刺版本 stats.ipynb
    - 未来使用 insight
